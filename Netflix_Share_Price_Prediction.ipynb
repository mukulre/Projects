```python
def time_series_plot(dataframe):
 '''
 This function generates a side-by-side comparison of time series
plots for the "Open" and "Close" columns.

 Args:
 dataframe: The DataFrame containing the time series data.

 Returns:
 None
 '''
 # Get the date column from the DataFrame
 x = dataframe['Date']

 # Get the values for the "Open" and "Close" columns
 open_prices = dataframe['Open']
 close_prices = dataframe['Close']

 # Create a figure and two subplots side by side
 fig, axs = plt.subplots(1, 2, figsize=(12, 6))
 # Plot the time series for the "Open" column
 axs[0].plot(x, open_prices, color='blue')
 axs[0].set_title('Open Prices')
 axs[0].set_xlabel('Date')
 axs[0].set_ylabel('Price')

 # Plot the time series for the "Close" column
 axs[1].plot(x, close_prices, color='red')
 axs[1].set_title('Close Prices')
 axs[1].set_xlabel('Date')
 axs[1].set_ylabel('Price')

 # Adjusting layout to prevent overlap of labels
 plt.tight_layout()
 plt.show()
# Call the function with your DataFrame
time_series_plot(netflix)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, accuracy_score
import numpy as np

# Load the dataset
netflix = pd.read_csv('NFLX.csv')

# Prepare input features and target variable for regression (Linear Regression)
X_reg = netflix[['Open', 'High', 'Low', 'Volume']]
y_reg = netflix['Close']  # Continuous target variable

# Prepare input features and target variable for classification (Logistic Regression and KNN)
X_clf = netflix[['Open', 'High', 'Low', 'Volume']]
y_clf = (netflix['Close'] > netflix['Close'].median()).astype(int)  # Binary target variable

# Split the dataset into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Split the dataset into training and testing sets for classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train_reg, y_train_reg)
y_pred_reg = linear_model.predict(X_test_reg)
mse_linear = mean_squared_error(y_test_reg, y_pred_reg)

# Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(X_train_clf, y_train_clf)
y_pred_clf_logistic = logistic_model.predict(X_test_clf)
accuracy_logistic = accuracy_score(y_test_clf, y_pred_clf_logistic)

# K-Nearest Neighbors Model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_clf, y_train_clf)
y_pred_clf_knn = knn_model.predict(X_test_clf)
accuracy_knn = accuracy_score(y_test_clf, y_pred_clf_knn)

# Print results
print("Linear Regression MSE:", mse_linear)
print("Logistic Regression Accuracy:", accuracy_logistic)
print("KNN Accuracy:", accuracy_knn)

# Determine the best model based on performance metrics
best_model = "Linear Regression" if mse_linear < min(1 - accuracy_logistic, 1 - accuracy_knn) else "Classification Models"
print("Best Model:", best_model)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, accuracy_score

# Load the dataset
netflix = pd.read_csv('NFLX.csv')

# Prepare input features and targets
X = netflix[['Open', 'High', 'Low', 'Volume']]
y_reg = netflix['Close']  # Continuous target for Linear Regression
y_clf = (netflix['Close'] > netflix['Close'].median()).astype(int)  # Binary target for classification

# Split the dataset into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# Split the dataset into training and testing sets for classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X, y_clf, test_size=0.2, random_state=42)

# Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train_reg, y_train_reg)
y_pred_reg = linear_model.predict(X_test_reg)
mse_linear = mean_squared_error(y_test_reg, y_pred_reg)

# Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(X_train_clf, y_train_clf)
y_pred_clf_logistic = logistic_model.predict(X_test_clf)
accuracy_logistic = accuracy_score(y_test_clf, y_pred_clf_logistic)

# K-Nearest Neighbors Model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_clf, y_train_clf)
y_pred_clf_knn = knn_model.predict(X_test_clf)
accuracy_knn = accuracy_score(y_test_clf, y_pred_clf_knn)

# Print results
print("Linear Regression MSE:", mse_linear)
print("Logistic Regression Accuracy:", accuracy_logistic)
print("KNN Accuracy:", accuracy_knn)

# Determine the best model based on performance metrics
best_accuracy = max(accuracy_logistic, accuracy_knn)
best_model = "Logistic Regression" if best_accuracy == accuracy_logistic else "KNN"
print("Best Model for Accuracy:", best_model)
```

```python
import pandas as pd
from transformers import pipeline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load the dataset
file_path = '/content/NFLX.csv'
df = pd.read_csv(file_path)

# Display the first few rows to understand the data structure
print(df.head())

# Assuming the dataset contains a column 'text' with textual data
if 'text' in df.columns:
    # Split the data into training and testing sets
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

    # Initialize transformer models for sentiment analysis
    models = {
        "DistilBERT": pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english"),
        "RoBERTa": pipeline("sentiment-analysis", model="roberta-base"),
        "BART": pipeline("sentiment-analysis", model="facebook/bart-large-imdb")
    }

    # Function to get predictions
    def get_predictions(model, texts):
        return [model(text)[0]['label'] for text in texts]

    # Dictionary to store results
    results = {}

    # Apply each model and calculate accuracy
    for model_name, model in models.items():
        # Get predictions for the test set
        predictions = get_predictions(model, test_df['text'].tolist())

        # Calculate accuracy (assuming binary labels: 'POSITIVE' and 'NEGATIVE')
        accuracy = accuracy_score(test_df['sentiment'].tolist(), predictions)
        results[model_name] = accuracy

    # Display results
    for model_name, accuracy in results.items():
        print(f"{model_name} Accuracy: {accuracy:.2f}")

    # Determine the best model based on accuracy
    best_model = max(results, key=results.get)
    print(f"Best Model: {best_model} with Accuracy: {results[best_model]:.2f}")

# Save the results to a new CSV file if needed
output_path = '/content/NFLX_with_sentiment_comparison.csv'
df.to_csv(output_path, index=False)
print(f"Results saved to {output_path}")
```
