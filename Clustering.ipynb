Scenario:
A retail store wants to understand its customers better by segmenting them into groups based on their purchasing habits. This segmentation will help the store target specific marketing strategies to different customer groups.

The dataset contains customer data with the following features:

Annual Income (in $1000s): How much the customer earns annually.
Spending Score (1–100): A measure of the customer's spending habits (higher is better).
We will use KMeans Clustering and Agglomerative Clustering to group customers into segments.

Sample Dataset
Here’s a small dataset of customers:

Customer ID	Annual Income ($1000s)	Spending Score (1-100)
1	15	39
2	16	81
3	17	6
4	18	77
5	19	40
6	20	76
7	21	6
8	22	94
9	23	3
10	24	72


```python
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt

# Sample data: Annual Income and Spending Score
data = np.array([
    [15, 39], [16, 81], [17, 6], [18, 77], [19, 40],
    [20, 76], [21, 6], [22, 94], [23, 3], [24, 72]
])

# Number of clusters
k = 3

# KMeans Clustering
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data)
kmeans_labels = kmeans.labels_
kmeans_centroids = kmeans.cluster_centers_

# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
agglo_labels = agglo.fit_predict(data)

# Plotting Results
plt.figure(figsize=(14, 6))

# KMeans Plot
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=kmeans_labels, cmap='viridis', s=100)
plt.scatter(kmeans_centroids[:, 0], kmeans_centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("KMeans Clustering")
plt.xlabel("Annual Income ($1000s)")
plt.ylabel("Spending Score (1-100)")
plt.legend()

# Agglomerative Clustering Plot
plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], c=agglo_labels, cmap='viridis', s=100)
plt.title("Agglomerative Clustering")
plt.xlabel("Annual Income ($1000s)")
plt.ylabel("Spending Score (1-100)")

plt.tight_layout()
plt.show()
```
