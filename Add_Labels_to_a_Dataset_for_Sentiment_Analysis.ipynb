```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
import pandas as pd

# Using the 'python' engine to handle potential quoting issues
data = pd.read_csv("/content/reviews data.csv", engine='python', on_bad_lines='warn')
# engine='python': Uses the Python parsing engine which might be more flexible.
# on_bad_lines='warn': Prints a warning for problematic lines instead of raising an error.

data = data.dropna()
print(data.head())
```

```python
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Review"]]
data['Compound'] = [sentiments.polarity_scores(i)["compound"] for i in data["Review"]]
data.head()
```

```python
score = data["Compound"].values
sentiment = []
for i in score:
    if i >= 0.05 :
        sentiment.append('Positive')
    elif i <= -0.05 :
        sentiment.append('Negative')
    else:
        sentiment.append('Neutral')
data["Sentiment"] = sentiment
data.head()
```

```python
print(data["Sentiment"].value_counts())
```

```python
data.to_csv("new_data.csv")
```
