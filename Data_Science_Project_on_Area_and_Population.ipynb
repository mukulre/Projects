```python
import pandas as pd
cities = pd.read_csv("california_cities.csv")
print(cities.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn

# Set the aesthetic style of the plots
seaborn.set()

# Verify the column names in the dataframe
print(cities.columns)

# Correct the column names if they are different
plt.figure(figsize=(10, 6))  # Optional: Set figure size for better visibility

# Create a scatter plot
scatter = plt.scatter(
    cities['longd'],
    cities['latd'],
    c=np.log10(cities['population_total']),
    cmap='viridis',
    s=cities['area_land_km2'],
    linewidth=0,
    alpha=0.5
)

# Set axis properties
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(scatter, label='log$_{10}$(population)')
plt.clim(3, 7)

# Create a legend for different area sizes
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + ' km$^2$')

plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.title("Area and Population of California Cities")
plt.show()
```
