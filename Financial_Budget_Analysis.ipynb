```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("India_budget_2021.csv")
data.head()
# Print the shape of the DataFrame to check number of rows and columns before slicing
print("Shape of DataFrame before slicing:", data.shape)

# Ensure your indices are within the valid range
# Replace with appropriate indices based on your DataFrame's size
data = data.iloc[[0, 1, 2, 3, 4, 5, 6, 7, 8],:]
#For example, if data has 10 rows, valid indices are 0 to 9.

# Print the shape of the sliced DataFrame
print("Shape of DataFrame after slicing:", data.shape)

row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
# Use pd.concat instead of append
data = pd.concat([data, pd.DataFrame([row])], ignore_index=True)
print(data)
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
